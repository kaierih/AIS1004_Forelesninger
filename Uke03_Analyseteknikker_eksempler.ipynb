# Eksempel Sløyfeanalyse:



## Ligningssett:
$$\begin{align}
7\Omega \cdot I_1 - 2 \Omega \cdot I_2 + 0\Omega \cdot I_3 &= 10V \\
-2\Omega \cdot I_1 + 5\Omega \cdot I_2 - 2 \Omega \cdot I_3 &= 0V \\
0\Omega \cdot I_1 - 2 \Omega \cdot I_2 + 9 \Omega \cdot I_3 &= 0V
\end{align}$$

```python
import numpy as np

A = np.array([[7, -2, 0],
             [-2, 5, -2],
              [0, -2, 9]])

b = np.array([[10],
              [0],
              [0]])

I = np.linalg.solve(A, b)
I
```

$$I_y = I_3$$

```python
Iy_a = float(I[2,0])
print(f"{Iy_a = :.3f}A")
```

# Eksempel Sløyfeanalyse:



## Ligningssett:
$$\begin{align}
\frac{17}{10} \cdot u_1 - 1 \cdot u_2 + 0\cdot u_3 &= 0V \\
-1 \cdot u_1 + \frac{11}{6} \cdot u_2 - \frac{1}{3} \cdot u_3 &= 0V \\
0\cdot u_1 - \frac{1}{3} \cdot u_2 + \frac{7}{12} \cdot u_3 &= -2V
\end{align}$$

```python
import numpy as np
A = np.array([[17/10, -1, 0],
              [-1, 11/6, -1/3],
              [0, -1/3, 7/12]])
b = np.array([[0],
              [0],
              [-2]])

u = np.linalg.solve(A, b)    # Løs Ligningssettet
u_1, u_2, u_3 = u.flatten()  # 
```

$$I_y = \frac{u_2 - u_3}{3\Omega}$$

```python
Iy_b = (u_2 - u_3)/3  # Ohms lov
print(f"{Iy_b = :.3f}A")    # Skriv ut svar
```
